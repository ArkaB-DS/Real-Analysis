# <font color=brown>Bisection Method of Finding Roots</font>
## <font color=green>This is based on **Location of Roots Theorem** </font>
Suppose I want to find out the solution of the equation: $xe^x=2$.
Take $f(x)=xe^x-2$.
Now, we first plot the function to find an interval where $f(x)$ changes sign.
Then we use the **Location of Roots Theorem** on that interval.

```r
x=seq(-1,1,length=1000) # change according to f(x)
f<-function(x) x*exp(x)-2 # change f(x) as needed
y=f(x)
par(bg="pink")
plot(x,y,type="l",col="green",lwd=2)
abline(h=0,col="red",lty=2,lwd=2)
# from the graph below it is evident that f(x) changes sign in the interval[-1/2,1/2].

# The Bisection Method
a=-1
b=1
c=(a+b)/2
epsilon=0.00000001 # error margin
while(abs(f(c))>=epsilon)
    {
    if (f(a)>0 & f(c)<0)
        {
        b=a
        a=c
    }
    else if (f(a)<0 & f(c)>0)
        {
        b=c
    }
    else if (f(b)>0 & f(c)<0)
        {
        a=c
    }
    else if (f(b)<0 & f(c)>0)
        {
        a=b
        b=c
    }
    c=(a+b)/2
}
cat("The root of the equation is: ",c)
```
