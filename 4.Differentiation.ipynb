### *Example of a Continuous EVERYWHERE but Differentiable NOWHERE Function*

```r
options(warn=-1)
f<-function(x) {
	z=sum(1/2^(1:100)*cos(3^(1:100)*x),na.rm=T)
	return (z) 
}

x<-seq(-2,2,length=1000)
y=sapply(x, function(x) f(x) )
library(ggplot2)
plt<-ggplot(data.frame(x=x,y=y),aes(x=x,y=y))
plt+geom_line(col="red4")+
	ggtitle("A continuous EVERYWHERE, but differentiable NOWHERE function")+
	theme(
		plot.title=element_text(color="seagreen",face="italic",hjust=0.5)
		)

	
```

```r
h <- function(x) return (exp(exp(x))) # enter first function
g <- function(x) return (5) # enter second function

library(ggplot2)
x<-seq(-1,1,length=1000)
plt<-ggplot(data.frame(x=x,y=sapply(x,function(x) h(x) )),aes(x=x,y=y))+
    geom_line(color="red")+  
    geom_line( aes( x=x, y=sapply ( x, function(x) g(x) )) , color="blue")                                      

f <- function(x) return ( h(x)-g(x) )
Df <- function(x) return ( exp(exp(x))*exp(x) )
z <- NULL
x0 <- 1 
x1 <- x0 - f(x0)/Df(x0)
z<-c(x0,x1)
epsilon <- 0.0001
while ( abs(x0-x1) > epsilon ) {
    x0 <- x1
    x1 <- x0 - f(x0)/Df(x0)
    z<-c(z,x1)
}
round (x1, 5)
plt+
    geom_vline(xintercept=x1,color="seagreen") +
    scale_color_manual(values=c("red","blue","seagreen"),
                       labels=c("h(x)","g(x)","Newton Raphson's solution"))
    # annotate("text",x=1.1,y=2.4,label="h(x)")+
    # annotate("text",x=1,y=0.75,label="g(x)")+
    # annotate("text",x=-0.60,y=2,label="Newton Raphson Solution")
```
