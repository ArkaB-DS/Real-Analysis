# <font color=red>Calculation of Square Roots</font>
## <font color=seagreen>This is based on 3.3.5 Example.</font>
#### <font color=blue>One can realize after going through the follwing code that it is basically an application of </font><font color=brown>_**Newton Raphson's Method**_</font>

```r
square_root<-function(a,epsilon) # a>0 is the number whose positive square root we intend to find epsilon=degree of accuracy
{
    if (a<0) print("Square root of negative numbers is not defined!") else {
    x1=1 # this will represent a sequence which will eventually give the result
    x2=1/2*(x1+a/x1)
    while (abs(x1-x2)>=epsilon)
        {
        x1=x2
        x2=1/2*(x1+a/x1)
    }}
    return (x2)
}

# Let's check the square root of 41 with 0.0001 as the margin of error
cat("The square root of 41 is: ",square_root(41,0.0001))
```

# <font color=red> An approximation of the Euler's constant "e"</font>

We know that we that  $$lim_{n\rightarrow{\infty}}(1+1/n)^{n}=e$$
We use the sequence $$e_n=(1+1/n)^n$$ to approximate the value of e. 

```r
e_n<-function(n)  ((1+1/n)^n)
i=1
while(abs(e_n(i)-e_n(i+1))>=0.0000000000001) i=i+1

cat("An approximate value of e with 0.0000000000001 error margin is:  ",e_n(i))
```

# <font color=red> The graph of divergent sequences <font color=green>cos(n)</font> and <font color=green>sin(n)</font></font>

```r
options(warn=-1)
x=seq(-2*pi,2*pi,length=1000)
y1=sin(x)
y2=cos(x)
plot(x,y1,col="magenta",main="Graph of cos(n) and sin(n)",xlab="x",ylab="",type="l",lwd=2)
par(new=T)
plot(x,y2,main="",xaxt="n",yaxt="n",xlab="",ylab="",col="seagreen",type="l",,lwd=2)
abline(h=1/2,col="blue",lwd=1,lty=2)
abline(h=-1/2,col="blue",lwd=1,lty=2)
```

# <font color=red>A Comparison Between <font color=green>log(n)</font> and <font color=green>$$\sum_{n}1/n$$</font>

```r
log_n<-function(n) log(n)
s_n<-function(n)   sum(1/(1:n))

n=1:100000

Log<-sapply(n,log_n)
s<-sapply(n,s_n)
plot(n,Log,col="red",type="l",lwd=2,main="A Comparison between log(n) and the Harmonic Series",xlab="n",ylab="")
lines(s,col="seagreen",lwd=2)
```
