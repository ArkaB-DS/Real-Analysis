# <font  color=seablue>Graphical view of certain interesting functions whose limit does not exist </font>

```r
x1=seq(-10,-0.5,length=1000)
x2=seq(0.5,10,length=1000)
x=c(x1,x2)
par(mfrow=c(2,1))
plot(x,exp(1/x),col="blue",xlab="x",ylab=expression(e^(1/x)),main=paste("Graph of ",expression(e^(1/x))))
abline(v=0,h=1,lty=2,lwd=2,col="magenta")
plot(x,1/(1+exp(1/x)),col="red",xlab="x",ylab=expression(1/(1+e^(1/x))),main=paste("Graph of ",expression(1/(1+e^(1/x)))))
abline(v=0,h=0.5,lty=2,lwd=2,col="seagreen")
```
